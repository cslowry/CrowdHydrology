```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # or yolov8s.pt, yolov8m.pt, etc.
```

```python
model.train(data='data.yaml', epochs=50, imgsz=640)
```

```python
metrics = model.val()
```

```python
print(metrics.box.map)
print(metrics.box.map50)
print(metrics.box.map75)
print(metrics.box.maps)
```

```python
predicts = model.predict('data/test/images')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# Reference: https://docs.ultralytic.com/modes/predict/#key-features-of-predict-mode

fig, axes = plt.subplots(4, 5, figsize=(15, 12))
for i, result in enumerate(predicts):
    # Get annotated image
    row = i // 5
    col = i % 5
    pred_img = result.plot()

    # Show with matplotlib
    axes[row, col].imshow(pred_img)
    axes[row, col].axis('off')
plt.show()
```

```python
model.export(format="onnx")  # export to ONNX format
```

# Analyzing Training Performance

```python
import pandas as pd
results = pd.read_csv("runs/detect/train10/results.csv") # view training labels
results.head()
```

```python
plt.plot(results['epoch'], results['train/box_loss'], label='Box Loss')
plt.plot(results['epoch'], results['train/dfl_loss'], label='Distribution Focal Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Over Epochs')
plt.show()
```

```python
plt.plot(results['epoch'], results['metrics/precision(B)'], label='Box Loss')
plt.xlabel('Epoch')
plt.ylabel('metrics/precision(B)')
plt.title('Box Precision Over Epochs')
plt.show()
```
